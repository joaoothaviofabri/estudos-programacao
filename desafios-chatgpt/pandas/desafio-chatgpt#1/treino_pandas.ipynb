## Ordenação e agregação

```python
# Ordenar usuários pelo login_count (decrescente)
df.sort_values(by="login_count", ascending=False)
```

```python
# Calcular a média de idade
df["age"].mean()
```

```python
# Calcular a média de logins por cidade
df.groupby("city")["login_count"].mean()
```

```python
# Descobrir qual usuário mais logou
df.loc[df["login_count"].idxmax()]
```

```python
# Descobrir qual cidade tem mais usuários
df["city"].value_counts()
```

## Cyber/Analytics 👾

```python
df_analise = df
display(df_analise)
```

```python
# Criar uma coluna chamada ativo: True se last_login_days <= 3, False caso contrário
df_analise["ativo"] = df_analise["last_login_days"] <= 3
display(df_analise)
```

```python
# Criar uma coluna nivel: "baixo" → login_count < 10, "medio" → entre 10 e 30, "alto" → acima de 30
df_analise["nivel"] = np.where(df_analise["login_count"] < 10, "baixo", np.where(df_analise["login_count"] <= 30, "medio", "alto"))
display(df_analise)
```

```python
# Listar apenas usuários ativos e com nível "alto"
df_analise.loc[df_analise["nivel"] == "alto"]
```

## Extra

```python
# Exportar um novo CSV só com usuários ativos
df_usuarios_ativos = df_analise.loc[df_analise["ativo"] == True]
display(df_usuarios_ativos)
```

```python
df_usuarios_ativos.to_csv("treino_pandas_usuarios_ativos.csv")
```
